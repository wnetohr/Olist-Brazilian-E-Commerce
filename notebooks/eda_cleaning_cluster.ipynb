```python
with open('./../data/clean_data/labels_encoders.json', encoding="utf-8") as json_file:
    labels_encoders = json.load(json_file)
    datas_comerciais = labels_encoders["commercial_date_coded"]
    janela_tempo_pedido = labels_encoders["time_window_order_coded"] 
    categorias = labels_encoders["filtered_category_coded"]
```

```python
df["filtered_category_coded"] = df["filtered_category_coded"].astype('str')
df["filtered_category"] = df["filtered_category_coded"].map(categorias)
df["filtered_category_coded"] = df["filtered_category_coded"].astype('int')

df["commercial_date_coded"] = df["commercial_date_coded"].astype('str')
df["commercial_date"] = df["commercial_date_coded"].map(datas_comerciais)
df["commercial_date_coded"] = df["commercial_date_coded"].astype('int')

df["time_window_order_coded"] = df["time_window_order_coded"].astype('str')
df["time_window_order"] = df["time_window_order_coded"].map(janela_tempo_pedido)
df["time_window_order_coded"] = df["time_window_order_coded"].astype('int')
df
```
